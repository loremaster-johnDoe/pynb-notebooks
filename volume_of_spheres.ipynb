***
$\mathbf{\text{Volume of an n-ball with radius R}}$<br>
***
$$ V_{n}(R) = \begin{cases}
1, & n=0 \\
2R, & n=1 \\
\frac{2\pi R^2}{n}.V_{n-2}(R), & otherwise
\end{cases}$$

```python
import math
```

```python
def vol(n, r=1):
    """
    Formula for the volume of an n-ball
    n : # of dimensions (∈ ℕ)
    r : radius of the n-ball
    """
    if n==0:
        return 1
    if n==1:
        return 2*r
    else:
        return 2*math.pi*r**2*vol(n-2, r)/n
```

```python
vol(2,1)
```

```python
x = [i for i in range(26)]
#v_5 = [vol(i, 0.5) for i in range(26)]
v_9 = [vol(i, 0.9) for i in range(26)]
v_10 = [vol(i, 1) for i in range(26)]
v_11 = [vol(i, 1.1) for i in range(26)]
#v_20 = [vol(i, 2) for i in range(26)]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))

colours = ['b', 'c', 'y', 'm', 'r']
#a = plt.scatter(x, v_5, color=colours[0])
b = plt.scatter(x, v_9, color=colours[1])
c = plt.scatter(x, v_10, color=colours[2])
d = plt.scatter(x, v_11, color=colours[3])
#e = plt.scatter(x, v_20, color=colours[4])

plt.legend((b,c,d), ('R = 0.9', 'R = 1', 'R = 1.1'),
          scatterpoints=1,
          loc='upper right',
          ncol=1,
          fontsize=10)
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
plt.scatter(x, v_10)
plt.title("Volume of an n-ball of unit radius (y-axis) vs n (no. of dimensions)")
plt.show()
```

Interestingly, the volume of a unit sphere (or even spheres with varying radii) first increases with the number of dimensions and then surprisingly diminishes to zero. 
Although, looking at the recursive sequence defined above, it is clear why this should be the case. However, intuitively it is quite confusing to have a sphere of sizeable radius having no volume in higher dimensions.
