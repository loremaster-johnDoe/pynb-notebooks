***
$\mathbf{\text{Plotting Primes in the Cartesian plane using Polar co-ordinates}}$<br>
***

```python
def is_prime(x):
    l = []
    if x == 1:
        return 0
    elif x == 2:
        return 1
    else:
        for i in range(2,x):
            if x%i==0:
                l.append(i)
        if len(l)>0:
            return 0
        else:
            return 1
```

```python
for i in range(11,31):
    print(f"{i} is prime : {is_prime(i)}")
```

```python
def get_primes(n):
    l=[]
    for i in range(1,n+1):
        if is_prime(i):
            l.append(i)
    return l
```

```python
get_primes(10)
```

The following code plots prime numbers using the following rule :
\begin{array}{l}
x = p*cos(p), & \mbox{p is prime} \\
y = p*sin(p), & \mbox{''    ''}
\end{array}

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
fig = plt.figure(figsize=(15,15))
r_theta = get_primes(10000)
x = [i*math.cos(i) for i in r_theta]
y = [i*math.sin(i) for i in r_theta]
plt.scatter(x,y)
plt.show()
```
